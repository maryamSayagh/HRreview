````python
#selecting only rows from 2010 to now
import re

def keep_journals_active_after(df, year_col='Coverage', min_year=2010):
    def extract_years(text):
        if not isinstance(text, str):
            return []
        years = set()
        # Normalize en dash to hyphen
        text = text.replace('–', '-')
        parts = [p.strip() for p in text.split(';')]
        for p in parts:
            if '-' in p:
                match = re.findall(r'(\d{4})-(\d{4})', p)
                if match:
                    start, end = map(int, match[0])
                    years.update(range(start, end + 1))
            else:
                match = re.findall(r'\d{4}', p)
                for y in match:
                    years.add(int(y))
        return years

    def is_recent(year_set):
        return any(y >= min_year for y in year_set)

    df = df.copy()
    df['__years'] = df[year_col].apply(extract_years)
    df = df[df['__years'].apply(is_recent)].drop(columns='__years')
    return df

sc=scopus_clean

sc=keep_journals_active_after(sc, year_col='Coverage', min_year=2010)
sc.head()
````

````python


def query_elsevier_to_dataframe(api_key, query_bis, count=25, max_results=200):
    base_url = "https://api.elsevier.com/content/search/scopus"
    headers = {
        "X-ELS-APIKey": api_key,
       # "Accept": "application/json"
    }

    results = []
    start = 0

    while start < max_results:
        params = {
            "query": query,
            "count": count,
            "start": start,
            "sort": "relevance"
        }

        response = requests.get(base_url, headers=headers, params=params)

        if response.status_code != 200:
            print(f"Error {response.status_code}: {response.text}")
            break

        data = response.json()
        entries = data.get("search-results", {}).get("entry", [])

        for entry in entries:
            print(entry)
            title = entry.get("dc:title", "")
            authors = entry.get("dc:creator", "") or entry.get("author", "")
            pub = entry.get("prism:publicationName", "")
            year = entry.get("prism:coverDate", "")[:4]
            doi = entry.get("prism:doi", "") or extract_link_doi(entry.get("prism:url", ""))
            link = entry.get("prism:url", "")
            citedby = entry.get("citedby-count", "")

            results.append({
                "Title": title,
                "Authors": authors,
                "Publication": pub,
                "Year": year,
                "DOI": doi,
                "Link": link,
                "Cited By": citedby
            })

        if not entries or len(entries) < count:
            break

        start += count

    return pd.DataFrame(results)


def extract_link_doi(url):
    """
    Extracts DOI from Scopus URL if DOI is not directly given.
    """
    if url and "doi.org" in url:
        return url.split("doi.org/")[-1]
    return ""
````
